## Model Inference

### Import Library

````python
import joblib
import numpy as np
import pandas as pd
````

### Load Model

````python
# Loading Model
with open('scaler.pkl', 'rb') as file_1:
  scaler = joblib.load(file_1)
  
with open('model.pkl', 'rb') as file_2:
  model = joblib.load(file_2)
  
with open('num_col.pkl', 'rb') as file_3:
  num_f = joblib.load(file_3)
  
with open('cat_col.pkl', 'rb') as file_4:
  cat_f = joblib.load(file_4)
````

### Get Data for Model Inference

````python
rng = np.random.default_rng(seed=2895)
stat_pay_sep = rng.integers(0, 4, size=15)
stat_pay_aug = rng.integers(0, 4, size=15)
stat_pay_jul = rng.integers(0, 4, size=15)
stat_pay_jun = rng.integers(0, 4, size=15)
stat_pay_may = rng.integers(0, 4, size=15)
stat_pay_apr = rng.integers(0, 4, size=15)
limit_balance = rng.integers(1, 80, size=15)*10000
data_inf = pd.DataFrame({'stat_pay_sep' : stat_pay_sep,
                         'stat_pay_aug' : stat_pay_aug,
                         'stat_pay_jul' : stat_pay_jul,
                         'stat_pay_jun' : stat_pay_jun,
                         'stat_pay_may' : stat_pay_may,
                         'stat_pay_apr' : stat_pay_apr,
                         'limit_balance' : limit_balance})
data_inf
````

### Inferencing

1. Scaling & Encoding

````python
# Scaling data inference
data_inf_scaled = scaler.transform(data_inf[num_f])
````

2. concat data inference

````python
# concat data inference
data_inf_fix = np.concatenate([data_inf_scaled, data_inf[cat_f]],axis=1)
````

3. Prediction

````python
#prediction price from data inference
data_inf['Fault Payment'] = model.predict(data_inf_fix)
data_inf
````
